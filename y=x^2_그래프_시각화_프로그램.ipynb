```
import pandas as pd
import matplotlib.pyplot as plt
import os

def plot_quadratic_graph(file_path):
    """
    CSV 파일을 읽어서 x, y 그래프를 그리는 함수
    """
    # 1. 파일 존재 여부 확인
    if not os.path.exists(file_path):
        print(f"Error: '{file_path}' 파일이 존재하지 않습니다. 파일을 먼저 업로드해주세요.")
        return

    try:
        # 2. CSV 데이터 읽기
        # x, y 두 개의 열로 구성된 데이터를 판다스로 읽어옵니다.
        df = pd.read_csv(file_path)

        # 데이터 확인 (상위 5개)
        print("데이터 로드 성공! 데이터 샘플:")
        print(df.head())

        # 3. 그래프 설정
        plt.figure(figsize=(10, 6))
        plt.plot(df['x'], df['y'], label='y = x^2', color='blue', linewidth=2)

        # 그래프 스타일링
        plt.title('Visualization of y = x^2 from CSV Data', fontsize=15)
        plt.xlabel('x value', fontsize=12)
        plt.ylabel('y value (x**2)', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()
        plt.axhline(0, color='black',linewidth=1) # x축 강조
        plt.axvline(0, color='black',linewidth=1) # y축 강조

        # 4. 결과 출력
        # 클라우드 환경(Colab 등)에서는 화면에 즉시 표시하고,
        # 일반 서버 환경에서는 파일로 저장합니다.
        output_image = 'graph_result.png'
        plt.savefig(output_image)
        print(f"\n그래프가 '{output_image}'로 저장되었습니다.")

        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    # 데이터 파일 경로 설정
    csv_file = 'data.csv'

    # 그래프 그리기 실행
    plot_quadratic_graph(csv_file)
```
